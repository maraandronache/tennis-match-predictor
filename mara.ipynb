# Dependencies

```python
from pathlib import Path
import pandas as pd
import numpy as np
```

# Data Reading

Load data

```python
def load_csvs(folder="data/raw", pattern="*.csv"):
    """Load all CSV files from a specified folder into a dictionary of DataFrames."""
    data_path = Path(folder)
    csv_files = data_path.glob(pattern)
    
    data_frames = {}
    for csv_file in csv_files:
        df_name = csv_file.stem  # Use the file name without extension as the key
        data_frames[df_name] = pd.read_csv(csv_file)
    
    return data_frames
```

```python
print(load_csvs())
```
