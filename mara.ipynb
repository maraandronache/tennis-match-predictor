# Dependencies

```python
from pathlib import Path
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Data Reading

Loading all CSV files

```python
def load_csvs(folder="data/raw", pattern="*.csv"):
    """Load all CSV files from a specified folder into a dictionary of DataFrames."""
    data_path = Path(folder)
    csv_files = data_path.glob(pattern)
    
    data_frames = {}
    for csv_file in csv_files:
        df_name = csv_file.stem  # Use the file name without extension as the key
        data_frames[df_name] = pd.read_csv(csv_file)
    
    return data_frames

data_dict = load_csvs()
print(f"Loaded {len(data_dict)} files.")
```

Quick overview

```python
df = data_dict.get("atp_matches_2024")

name = "atp_matches_2024"
print(f"\n{name}")
print(f"Rows: {df.shape[0]}, Columns: {df.shape[1]}")
display(df.head())
display(df.info())
display(df.describe())

# missing values 

missing = df.isna().mean().sort_values(ascending=False)
missing = missing[missing > 0] 
display(missing)

plt.figure(figsize=(10, 5))
sns.barplot(x=missing.index, y=missing.values)
plt.xticks(rotation=90)
plt.title("Missing values per column")


# examples of numerical distribution
num_cols = df.select_dtypes(include=[np.number]).columns.tolist()
relevant_num = [c for c in num_cols if c.lower() in ["minutes", "w_ace", "l_ace", "w_df", "l_df", "w_svpt", "l_svpt"]]

for col in relevant_num:
    plt.figure(figsize=(6, 4))
    sns.histplot(df[col].dropna(), bins=30, kde=True)
    plt.title(f"{name} - Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.grid(alpha=0.3)
    plt.tight_layout()
    plt.show()
    
# examples of categorical distribution
cat_cols = df.select_dtypes(include=["object"]).columns.tolist()
relevant_cat = [c for c in cat_cols if c.lower() in ["surface", "round", "tourney_name", "tourney_level"]]

if not relevant_cat:
    relevant_cat = cat_cols[:3]

for col in relevant_cat:
    vc = df[col].value_counts().head(10)
    plt.figure(figsize=(8, 4))
    sns.barplot(x=vc.index, y=vc.values)
    plt.title(f"{name} - {col} distribution (top 10)")
    plt.xticks(rotation=30, ha="right")
    plt.ylabel("Count")
    plt.tight_layout()
    plt.show()

```
